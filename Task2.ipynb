```python
dataset.columns =['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region','longitude','latitude']
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'],dayfirst=True)
```

```python
dataset['States']= dataset['States'].astype('category')
```

```python
dataset['Month'] = dataset['Date'].dt.month
```

```python
dataset['Month_int'] = dataset['Month'].apply(lambda x : int(x))
```

```python
dataset['Month_name'] =  dataset['Month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
dataset['Frequency'] = dataset['Frequency'].astype('category')
```

```python
dataset.drop(columns='Month',inplace=True)
dataset.head(5)
```

```python
dataset_stats = dataset[['Estimated Unemployment Rate',
       'Estimated Employed', 'Estimated Labour Participation Rate']]


round(dataset_stats.describe().T,2)
```

```python
region_stats = dataset.groupby(['Region'])[['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']].mean().reset_index()

region_stats = round(region_stats,2)


region_stats
```

```python
heat_maps = dataset[['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region','longitude']]

heat_maps = heat_maps.corr()

plt.figure(figsize=(10,8))
sns.set_context('notebook',font_scale=1)
sns.heatmap(heat_maps, annot=True,cmap='summer');
```

```python
fig = px.box(dataset,x='States',y='Estimated Unemployment Rate',color='States',title='Unemployment rate',template='plotly')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
plot_ump = dataset[['Estimated Unemployment Rate','States']]

dataset_unemp = plot_ump.groupby('States').mean().reset_index()

dataset_unemp = dataset_unemp.sort_values('Estimated Unemployment Rate')

fig = px.bar(dataset_unemp, x='States',y='Estimated Unemployment Rate',color='States',
            title='Average Unemployment Rate in each state',template='plotly')

fig.show()
```

```python
fig = px.bar(dataset, x='Region',y='Estimated Unemployment Rate',animation_frame = 'Month_name',color='States',
            title='Unemployment rate across region from Jan.2020 to Oct.2020', height=700,template='plotly')

fig.update_layout(xaxis={'categoryorder':'total descending'})

fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 2000

fig.show()
```

```python
unemplo_du = dataset[['States','Region','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']]

unemplo = unemplo_du.groupby(['Region','States'])['Estimated Unemployment Rate'].mean().reset_index()
```

```python
fig = px.sunburst(unemplo, path=['Region','States'], values='Estimated Unemployment Rate',
                  color_continuous_scale='Plasma',title= 'unemployment rate in each region and state',
                  height=650,template='ggplot2')


fig.show()
```
